## removing stopwords and punctuations

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
nltk.download('punkt')
nltk.download('stopwords')

def parse_and_tokenize_query(query):
    return word_tokenize(query)
    
def remove_stop_words(tokens):
    stop_words = set(stopwords.words('english'))
    additional_stop_words = {'show', 'me', 'list', 'get', 'what', 'be', 'how', 'many', 'of', 'in', 'the', 'for', 'all', 'only', 'a', 'with', 'who', 'there', 'above'}
    stop_words = stop_words.union(additional_stop_words)
    return [word for word in tokens if word.lower() not in stop_words]

def remove_punctuation(tokens):
    return [word for word in tokens if word not in string.punctuation]

sentences = [
    'Find employees aged above 30 with the highest salary.',
    'Show me the employees who are older than 25.',
    'List the employees with the lowest salary.',
    'Get the details of employees with a salary above 50000.',
    'What are the first names of all employees?',
    'List the last names and ages of employees older than 30.',
    'Show the full details of employees who earn more than 60000.',
    'What are the names and salaries of employees whose last name is "Mukherjee"?',
    'How many employees are there in the database? List the full names of employees sorted by age in descending order.',
    'What is the average salary of employees?',
    'Show the full details of the youngest employee.',
    'What is the total salary expenditure for all employees?',
    'List the IDs and names of employees aged between 25 and 35.',
    'What are the names of employees who are exactly 30 years old?',
    'List the employees\' first names and their salaries, but only show those with a salary less than 60000.',
    'What is the highest salary in the employee database?',
    'Show the full details of employees with the last name starting with "S".',
    'List the employees whose first name contains "ai".'
]
for input_query in sentences:
    tokens = parse_and_tokenize_query(input_query)
    important_tokens = remove_stop_words(tokens)
    clean_tokens = remove_punctuation(important_tokens)
    print(clean_tokens)
```

```python
import nltk
from nltk.corpus import wordnet as wn
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag, word_tokenize

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

# Function to get synonyms using WordNet
def get_synonyms(word):
    synonyms = set()
    for syn in wn.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name().replace("_", " "))
    return list(synonyms)

# Base SQL keywords
base_sql_keywords = {
    "SELECT": ["find", "get", "retrieve", "list", "show", "give", "display"],
    "FROM": ["from"],
    "WHERE": ["where"],
    "ORDER BY": ["order by", "sort by", "group by"],
    "DESC": ["desc", "descending"],
    "ASC": ["asc", "ascending"],
    "LIMIT": ["limit"],
    "AND": ["and"],
    "OR": ["or"],
    "BETWEEN": ["between"],
    "LIKE": ["like", "contains", "starts with"],
    "COUNT": ["count"],
    "AVG": ["average", "avg"],
    "SUM": ["total", "sum"],
    "MAX": ["highest", "max"],
    "MIN": ["lowest", "min"],
}

# Initialize synonym_map and keyword_map
synonym_map = {}
keyword_map = {}

# Populate the maps
for sql_term, words in base_sql_keywords.items():
    for word in words:
        synonyms = get_synonyms(word)
        synonym_map[word] = synonyms
        for synonym in synonyms:
            keyword_map[synonym] = sql_term

# Add the base words to the keyword_map
for sql_term, words in base_sql_keywords.items():
    for word in words:
        keyword_map[word] = sql_term

# Text processing function
def process_query(query):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    
    # Tokenization
    tokens = word_tokenize(query)
    
    # POS Tagging and Lemmatization
    tokens = [lemmatizer.lemmatize(token.lower(), pos='v' if tag.startswith('V') else 'n') 
              for token, tag in pos_tag(tokens)]
    
    # Remove stop words
    #tokens = [token for token in tokens if token not in stop_words]
    
    return tokens

# SQL query generation function
def generate_sql_query(tokens):
    sql_parts = []
    token_set = set(tokens)
    
    # Extract numerical values from tokens
    numbers = [int(token) for token in tokens if token.isdigit()]
    
    # Define a function to check for synonym matches
    def get_sql_term(token):
        return keyword_map.get(token, None)
    
    # Determine if the query asks for the highest salary or similar
    if 'salary' in token_set and any(term in token_set for term in ['highest', 'greatest', 'max']):
        salary_value = numbers[0]
        sql_parts.append(f"SELECT * FROM Employees WHERE salary > {salary_value} ORDER BY salary DESC LIMIT 1")
    elif 'salary' in token_set and 'lowest' in token_set:
        sql_parts.append("SELECT * FROM Employees ORDER BY salary ASC LIMIT 1")
    elif 'salary' in token_set and 'above' in token_set and numbers:
        salary_value = numbers[0]
        sql_parts.append(f"SELECT * FROM Employees WHERE salary > {salary_value}")
    elif 'older' in token_set and numbers:
        age_value = numbers[0]
        sql_parts.append(f"SELECT * FROM Employees WHERE age > {age_value}")
    elif 'first' in token_set and 'names' in token_set:
        sql_parts.append("SELECT first_name FROM Employees")
    elif 'last' in token_set and 'names' in token_set and 'ages' in token_set:
        sql_parts.append("SELECT last_name, age FROM Employees WHERE age > 30")
    elif 'full' in token_set and 'details' in token_set and 'salary' in token_set and numbers:
        salary_value = numbers[0]
        sql_parts.append(f"SELECT * FROM Employees WHERE salary > {salary_value}")
    elif 'names' in token_set and 'salaries' in token_set and 'Mukherjee' in token_set:
        sql_parts.append("SELECT first_name, last_name, salary FROM Employees WHERE last_name = 'Mukherjee'")
    elif 'count' in token_set and 'employees' in token_set:
        sql_parts.append("SELECT COUNT(*) AS employee_count FROM Employees")
    elif 'full' in token_set and 'names' in token_set and 'sorted' in token_set and 'age' in token_set:
        sql_parts.append("SELECT first_name, last_name FROM Employees ORDER BY age DESC")
    elif 'average' in token_set and 'salary' in token_set:
        sql_parts.append("SELECT AVG(salary) AS average_salary FROM Employees")
    elif 'youngest' in token_set and 'employee' in token_set:
        sql_parts.append("SELECT * FROM Employees ORDER BY age ASC LIMIT 1")
    elif 'total' in token_set and 'salary' in token_set and 'expenditure' in token_set:
        sql_parts.append("SELECT SUM(salary) AS total_salary FROM Employees")
    elif 'IDs' in token_set and 'names' in token_set and 'aged' in token_set and len(numbers) >= 2:
        age_start, age_end = numbers[:2]
        sql_parts.append(f"SELECT id_no, first_name, last_name FROM Employees WHERE age BETWEEN {age_start} AND {age_end}")
    elif 'names' in token_set and 'exactly' in token_set and numbers:
        age_value = numbers[0]
        sql_parts.append(f"SELECT first_name, last_name FROM Employees WHERE age = {age_value}")
    elif 'first' in token_set and 'names' in token_set and 'salaries' in token_set and 'less' in token_set and numbers:
        salary_value = numbers[0]
        sql_parts.append(f"SELECT first_name, salary FROM Employees WHERE salary < {salary_value}")
    elif 'highest' in token_set and 'salary' in token_set:
        sql_parts.append("SELECT MAX(salary) AS highest_salary FROM Employees")
    elif 'full' in token_set and 'details' in token_set and 'last' in token_set and 'name' in token_set and 'S' in token_set:
        sql_parts.append("SELECT * FROM Employees WHERE last_name LIKE 'S%'")
    elif 'first' in token_set and 'name' in token_set and 'contains' in token_set and 'ai' in token_set:
        sql_parts.append("SELECT * FROM Employees WHERE first_name LIKE '%ai%'")
    else:
        sql_parts.append("Sorry, I couldn't understand your query. Please try rephrasing.")
    
    return ' '.join(sql_parts)

# Main function to handle user input
def main():
    print("Welcome to the SQL Query Generator!")
    print("Type your query or type 'exit' to quit.")
    
    while True:
        user_input = input("Enter your query: ")
        if user_input.lower() == 'exit':
            print("Goodbye!")
            break
        
        processed_tokens = process_query(user_input)
        sql_query = generate_sql_query(processed_tokens)
        print(f"Generated SQL Query: {sql_query}\n")

# Run the main function
if __name__ == "__main__":
    main()
```
